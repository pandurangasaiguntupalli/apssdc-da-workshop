```python
x2 = np.linspace(10,30,50)
x2
```

```python
plt.plot(x2, np.sin(x2))
plt.plot(x2, np.cos(x2))
```

```python
plt.plot(x2, np.sin(x2), color = 'yellow' , marker = '+')
plt.plot(x2, np.cos(x2), color = 'blue' , marker = 'd')
```

```python
plt.plot(x2, np.sin(x2), color = 'yellow' , marker = '+', markersize = 10)
plt.plot(x2, np.cos(x2), color = 'blue' , marker = 'd', markersize = 10, markerfacecolor='w')
```
